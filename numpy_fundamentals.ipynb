***Why NumPy Is Important for Handling Data***

Python is easy to read and write, but it is not optimized for heavy numerical computation using normal lists and loops.

***NumPy solves this problem.***

It allows us to:
- Store data in efficient multi-dimensional arrays
- Perform fast mathematical operations without explicit loops
- Work with structured data like vectors, matrices, and tensors
- Write clean and readable code for complex numerical tasks

In simple terms:  
**NumPy helps Python handle data efficiently and at high speed.**

```python
# Simple speed comparison: Python list vs NumPy array
# Goal: show why NumPy is faster than list

import time
import numpy as np

# Creating a normal Python list
my_list = list(range(1000000))

# Creating a NumPy array with the same data
my_array = np.arange(1000000)


# -------- Python List Timing --------
start = time.time()

# Add 1 to each element using a loop
for i in range(len(my_list)):
    my_list[i] = my_list[i] + 1

end = time.time()
list_time = end - start

print("Python list time:", list_time)


# -------- NumPy Array Timing --------
start = time.time()

# Add 1 to all elements at once
my_array = my_array + 1

end = time.time()
array_time = end - start

print("NumPy array time:", array_time)


# -------- Result --------
print("\nResult:")
print("List took :", list_time, "seconds")
print("Array took:", array_time, "seconds")
```

***What to Take Away from This Test***

Don’t worry about the small details of the code. Just remember this:

NumPy arrays are much faster than Python lists when doing math on lots of numbers.

***Why?***

Python lists work one number at a time, which is slow

NumPy arrays work on all numbers at once, which is fast

Both did the same job, but NumPy finished way quicker

**1. Scalar**

```python
# we create a array called array1 using np.array, of one dimension (scalar)
array1 = np.array(6)
```

```python
print("array1 : ",array1 )
print("dimension of array1 : ",array1.ndim)
print("shape of array1 : ",array1.shape)
print("dtype of array1 : ",array1.dtype)
print("size of array1 : ",array1.size)
```

lets analyze our output:

**array1** : 6

we created a NumPy array like np.array(6)
A single number, not a list or list of lists.

**dimension of array1**  : 0 (ndim)

0 dimensions ,called a scalar array.
It has no rows or columns, just a single value.

**shape of array1** : ()

Empty tuple () , confirms it has no axes.

**dtype of array1** : int64

The element type is integer (64-bit).

**size of array1** : 1

Total elements = 1 makes sense, because it’s just one number.

**2.Vectors**

```python
array2 = np.array([7 , 33 , 45])
```

```python
print("array2 : ",array2 )
print("dimension of array2 : ",array2.ndim)
print("shape of array2 : ",array2.shape)
print("dtype of array2 : ",array2.dtype)
print("size of array2 : ",array2.size)
```

lets analyze our output

**array2** : [ 7 33 45 ]

Created using np.array([7, 33, 45]).
This is a list of numbers, so NumPy treats it as a 1D array.

**dimension of array2** : 1 (ndim)

1 dimension, meaning it has a single axis.
Think of it as a row of numbers.

**shape of array2** : (3,)

The tuple (3,) shows the array has 1 axis with 3 elements.

**dtype of array2** : int64

All elements are integers (64-bit).

**size of array2**: 3

Total elements are 3, which matches the three numbers in the array.

**3.Matrices**

```python
array3 = np.array( [ [ 1,2] ,[8,9] ] )
```

```python
print("array3 : ",array3 )
print("dimension of array3 : ",array3.ndim)
print("shape of array3 : ",array3.shape)
print("dtype of array3 : ",array3.dtype)
print("size of array3 : ",array3.size)
```

lets analyze our output

**array3** : [[1 2] [8 9]]

Created using np.array([[1, 2], [8, 9]]).
This is a list of lists, so NumPy creates a 2D array.

**dimension of array3** : 2 (ndim)

2 dimensions mean the array has two axes.
One axis represents rows, the other represents columns.

**shape of array3**  : (2, 2)

The shape shows 2 rows and 2 columns.
First value 2 is the number of rows.
Second value 2 is the number of columns.

**dtype of array3 **: int64

All elements are integers (64-bit).

**size of array3** : 4

Total elements are 4.
Calculated as rows × columns = 2 × 2.

**4.Tensors**

```python
array4 = np.array( [ [ [1,2],[3,4] ] , [ [9,8] , [7,6]] ] )
```

```python
print("array4 : ",array4 )
print("dimension of array4 : ",array4.ndim)
print("shape of array4 : ",array4.shape)
print("dtype of array4 : ",array4.dtype)
print("size of array4 : ",array4.size)
```

lets analyze our output

**array4** : [[[1 2] [3 4]] [[9 8] [7 6]]]

Created using np.array([[[1, 2], [3, 4]], [[9, 8], [7, 6]]]).
This is a list of 2D arrays, so NumPy creates a 3D array.

**dimension of array4**: 3 (ndim)

3 dimensions mean the array has three axes.
You can think of it as two 2D matrices stacked together.

**shape of array4**: (2, 2, 2)

First 2 means there are 2 blocks (or layers).
Second 2 means each block has 2 rows.
Third 2 means each row has 2 columns.

**dtype of array4**: int64

All elements are integers (64-bit).

**size of array4** : 8

Total elements are 8.
Calculated as 2 × 2 × 2.
